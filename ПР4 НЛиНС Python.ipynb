# Задание4
Пусть у нас есть список монеток в кармане pocket = [1, 5, 10, 25, 50, 100] и пользователь который вводит номинал, если номинал больше 100, то мы добавим его в список, если меньше и не кратен 2, то отдадимпервую монетку, если меньше и кратен 2, то последнюю из списка.

```python
x = int(input('Введите номинал: '))
pocket = [1, 5, 10, 25, 50, 100]
```

```python
if x>100:
    pocket.append(x)
elif x%2 != 0 and x!=100:
    pocket.pop(0)
elif x%2 == 0 and x!=100:
    pocket.pop(-1)
pocket
```

# Задание5
Есть список овощей veg = ["томат", "свекла", "морковь", "картошка", "лук", "репа"]
Написать программу которая будет давать пользователю 3 попытки угадать овощи с помощью ввода.Подсказка: нужно будет использовать цикл с поиском внутри цикла с вводом строки пользователем

```python
veg = ["томат", "свекла", "морковь", "картошка", "лук", "репа"]
```

```python
for i in range(3):
    print('\n\nПопытка', i+1)
    q = input('\nВведите название овоща: ')
    if q in veg:
        print('Поздравляем! Вы угадали!')
        break
    else:
        print('\nПробуйте еще раз')
        continue
```

# Домашнее задание
Написать программу которая будет составлять список фильмов, пользователь может удалять, добавлять,изменять или просматривать элементы из списка, также должен быть доступна возможность выхода изцикла с помощью оператора break. Для того чтобы цикл был бесконечным количество итераций устанавливаем на 9999

```python
data = []
```

```python

for i in range(9999):
    com = input('Введите команду \n"+" - Добавить Фильм\n"-" - Удалить фильм\n"/" - Изменить фильм\n"!" просмотреть Список\n\n" " - Выйти\n\n\n')
    if com == ' ':
        break
    elif com == '+':
        title = input('Введите название фильма: ')
        data.append(title)
        print('Фильм успешно добавлен!\n')
    elif com == '-':
        data.remove(title)
    elif com == '/':
        print(data)
        chid = int(input('Введите индекс элемента который вы хотите изменить: '))
        data.pop(chid)
        title = input('Введите название фильма: ')
        data.append(title)
        print(data)
    elif com == '!':
        print(data)
    
    
    
```
