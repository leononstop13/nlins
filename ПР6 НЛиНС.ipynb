```python
from keras.models import Sequential 
from keras.layers import Dense 
from keras.utils import np_utils
```

```python
wine=pd.read_csv('/Users/leonid/Downloads/Wine.txt',sep='\t',header=0)
wine.head()
```

```python
#смотрим соотношение элементов в последнем стобце
wine['Desired1(3)'].value_counts(normalize=True)
```

```python
#  предикторы и отклик разделяем
# Отклик - группирующая переменная -  вектор y
y=wine['Desired1(3)']
# Предикторы - таблица X
X=wine.drop('Desired1(3)',axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=1234)
```

```python
#Cмотрим количество элементов в каждой выборке

#X_train и y_train - обучаем на этих выборках
#X_test и y_test - тестируем точность на тренированной модели на этой выборке

X_train.head()
```

```python
X_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
#  Преобразование pandas dataframe в numpy array
X_train=X_train.values
X_test=X_test.values
y_train=y_train.values
y_test=y_test.values
```

```python
# Больше 2-х классов
# Классы не упорядочены
# np_utils делает преобразование в бинарную матрицу
y_train_bin=np_utils.to_categorical(y_train)

y_test_bin=np_utils.to_categorical(y_test)
```

```python
# Почемунеработает? - !Это не датафрейм
y_train_bin.head()
```

```python
#После конвертации в тензоры, классы 0,1,2 стали набором
# нолей и единиц, 1,0,0 - 0-ойкласс, 0,1,0 - 2й, 0,0,1 - 3-й    
y_train_bin[0:5]
```

```python
# Создание модели и указание параметров
model=Sequential()
#Добавляемслой 9 нейронов, 13 параметровнавход, используемакт. функ. relu
model.add(Dense(9,input_dim=13,activation='relu'))
#на скрытых слоях мы не указываем сколько параметров подается на вход, используем softmax
model.add(Dense(10,activation='relu',))
model.add(Dense(3,activation='softmax'))

# Построениемодели (методминим., критерийкач., метрика)
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])

# Запусктренировкимодели, указываемколичествоэпоха
model.fit(X_train,y_train_bin,epochs=300,batch_size=10)
```

```python
# Оценкамодели
scores=model.evaluate(X_test,y_test_bin)
print("\nТочность: %.2f%%"%(scores[1]*100))
```

```python
# Вычисляемпредсказания. смотрим какие бы получились y-ки по X-ам
predictions=model.predict(X_test)
print(predictions[0:5])
```
